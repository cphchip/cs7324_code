```python
import numpy as np

# Define a simple quadratic cost function (for demonstration)
def cost_function(x):
    return x ** 2

# Gradient of the cost function
def gradient(x):
    return 2 * x

# Parameters
learning_rate = 0.1
momentum = 0.9  # Momentum term
num_iterations = 50

# Initialize variables
x = 4.0  # Initial parameter value
velocity = 0.0  # Initial velocity

# Lists to store parameter values and costs during optimization
param_history = []
cost_history = []

# Gradient descent with momentum
for i in range(num_iterations):
    grad = gradient(x)
    velocity = momentum * velocity - learning_rate * grad
    x = x + velocity

    # Store parameter and cost values for visualization
    param_history.append(x)
    cost_history.append(cost_function(x))

    print(f"Iteration {i + 1}: x = {x:.4f}, Cost = {cost_function(x):.4f}")

# Plotting the optimization path
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(param_history, cost_history, marker='o')
plt.title("Optimization Path")
plt.xlabel("Parameter Value (x)")
plt.ylabel("Cost")

plt.subplot(1, 2, 2)
plt.plot(range(num_iterations), param_history, marker='o')
plt.title("Parameter Value Over Iterations")
plt.xlabel("Iteration")
plt.ylabel("Parameter Value (x)")

plt.show()
```
